### 문자열, 정수와 실수

위 코드를 다시 살짝 수정해 볼까요?

```python
a='나는 단팥빵'
bread=2
b='개를 먹었다.'
print(a,bread+1,b)
```

숫자 뿐 아니라 문장 역시 꼭 print ( ) 함수 내에 적지 않아도 됩니다. 작은 따옴표로 묶인 '나는 단팥빵', '개를 먹었다' 는 문자열String 이라고 부릅니다. 또한 프로그래밍에서 2, 3 과 같은 숫자는 정수int , 2.1, 3.3 과 같은 숫자는 실수float 라고 부릅니다.

### 벡터를 표현하려면?

파이썬 프로그래밍에서는 벡터를 다양한 방법으로 표현할 수 있습니다. 본 강좌에서는 **리스트***list* 라는 것을 활용합니다. 3차원 벡터 (1, 2, 1)를 한 번 만들어볼까요? 벡터를 만들고 print( ) 함수로 출력까지 해보겠습니다.

```python
a=[1, 2, 1]
print(a)
```

이렇듯 아주 손쉽게 우리는 다차원의 벡터를 만들어 낼 수가 있습니다. 

이번에는 문자열로 이루어진 리스트를 만들어 보겠습니다. food라는 리스트에 '치킨', '짜장면', '도시락', '탕수육' 이라는 음식이름 문자열을 넣고, 첫번째 음식인 '치킨'을 꺼내보겠습니다.

```python
food=['치킨','짜장면','도시락','탕수육']
print('나는', food[1], '이 먹고 싶어요') 
```

앗! 분명히 '치킨'을 출력하기 위해 food[1] 이라고 적었는데, 두번째 음식인 '짜장면'이 출력 되었네요. (대부분의 컴퓨터 프로그램에서 첫번째 항목이라는 것을 0번째로 표현합니다. 여기서도 food[1]이 첫번째 항목이 아니고, food[0]이 첫번째 항목임에 주의하세요.) 그럼 이에 맞게 코드를 수정해 보겠습니다.

```python
food=['치킨','짜장면','도시락','탕수육']
print('나는', food[0], '이 먹고 싶어요')
```
